```python
#Libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OneHotEncoder

import pickle

from sklearn.metrics import accuracy_score
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
spaceship_clean = spaceship.dropna()
spaceship_clean.info()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
#spaceship_clean['Cabin'] = spaceship_clean['Cabin'].str[0]
spaceship_clean.loc[:, 'Cabin'] = spaceship_clean.loc[:,'Cabin'].str[0]
```

```python
# We keep only the relevant categories
spaceship_clean['Cabin'] = spaceship_clean['Cabin'].apply(lambda x: x if x in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'] else 'U')
```

- Drop PassengerId and Name

```python
#your code here
spaceship_clean = spaceship_clean.drop(columns=['Name'], axis=1)
```

```python
spaceship_clean = spaceship_clean.drop(columns=['PassengerId'], axis=1)
```

```python
spaceship_clean.info()
```

- For non-numerical columns, do dummies.

```python
X = spaceship_clean.drop(columns=['Transported'], axis=1)
X
```

```python
y = spaceship_clean['Transported']
y
```

**Perform Train Test Split**

```python
#your code here
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

```python
X_train_num = X_train.select_dtypes('number')
X_test_num  = X_test.select_dtypes('number')

X_train_cat = X_train.select_dtypes('object')
X_test_cat  = X_test.select_dtypes('object')
```

```python
ohe = OneHotEncoder(sparse_output=False)

ohe.fit(X_train_cat)
X_train_cat_trans_np = ohe.transform(X_train_cat)
X_train_cat_trans_np
```

```python
X_train_cat_trans_df = pd.DataFrame(X_train_cat_trans_np, columns=ohe.get_feature_names_out(), index=X_train.index)
X_train_cat_trans_df
```

```python
X_test_cat_trans_np = ohe.transform(X_test_cat)
X_test_cat_trans_np
```

```python
X_test_cat_trans_df = pd.DataFrame(X_test_cat_trans_np, columns=ohe.get_feature_names_out(), index=X_test.index)
X_test_cat_trans_df
```

```python
# Concatenate numerical and categorical features for training set
X_train_final = pd.concat([X_train_num, X_train_cat_trans_df], axis=1)

# Concatenate numerical and categorical features for test set
X_test_final = pd.concat([X_test_num, X_test_cat_trans_df], axis=1)
```

```python
# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training data
X_train_scaled = scaler.fit_transform(X_train_final)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test_final)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here
# 1) Merge numerical with the new categorical columns (transformar las variables para que todas las columnas tengan la misma escala)
# 2) Transform to MinMax and Std Scaler 
# Class Pipelines (column transformer) can be saved with Pickle
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
y_pred = knn.predict(X_test_scaled)
y_pred
```

- Evaluate your model's performance. Comment it

```python
#your code here

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Evaluate the model on the training set
print(f"The accuracy of the model on the TRAINING set is: {knn.score(X_train_scaled, y_train)*100: .2f}%")

# Evaluate the model on the test set
print(f"The accuracy of the model on the TEST set is: {knn.score(X_test_scaled, y_test)*100: .2f}%")
```

### The accuracy of the model is 76.25%
